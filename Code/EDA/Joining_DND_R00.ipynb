We have a problem. The experiments used a different instrument to measure the personality.
Researching about this I realized that the minimum and maximum values for every dimesion is different depending on the test used.

                (min, max)

`R00 -> FFI ->  (12, 60)`

`DND -> PIR3 -> (48, 240)`

We could scale it using:

$$\dfrac{x_i - min(x)}{max(x) - min(x)}$$

```python
cols_to_modif = ['Neuroticism', 'Extraversion', 'Conscientiousness']
new_cols = ['Neuroticism_scaled', 'Extraversion_scaled', 'Conscientiousness_scaled']

def min_max(x, minimum, maximum):
    '''https://en.wikipedia.org/wiki/Feature_scaling'''
    return (x - minimum) / (maximum - minimum)
```

```python
#This is not maybe the ideal option but it gives similar values
min_max(dnd_survey['NEO-SF']['Conscientiousness'], minimum = 48, maximum = 240).mean() , min_max( r00_survey['NEO-SF']['Conscientiousness'], minimum = 12, maximum = 60).mean()
```

```python
min_max(dnd_survey['NEO-SF']['Neuroticism'], minimum = 48, maximum = 240).mean() , min_max(r00_survey['NEO-SF']['Neuroticism'], minimum = 12, maximum = 60).mean()
```

```python
min_max(dnd_survey['NEO-SF']['Extraversion'], minimum = 48, maximum = 240).mean() , min_max( r00_survey['NEO-SF']['Extraversion'], minimum = 12, maximum = 60).mean()
```

```python
#R00
for col,new_col in zip(cols_to_modif,new_cols):
    r00_survey['NEO-SF'][new_col] =  min_max(r00_survey['NEO-SF'][col], minimum = 12, maximum = 60)
    
```

```python
#DND
for col,new_col in zip(cols_to_modif,new_cols):
    dnd_survey['NEO-SF'][new_col] =  min_max(dnd_survey['NEO-SF'][col], minimum = 48, maximum = 240)
```

```python
NEO_SF =  pd.concat( [r00_survey['NEO-SF'] , dnd_survey['NEO-SF']], sort = False)
```

```python
import pandas as pd
```

```python
ids = [1020,1030,1008,1051,1097,1096,1041,1014,1029,1035,1049,1084,1032,1073,1039,1087,1021,1004,1050,1024,1018,1019,1085,1009,1080,1016,1002,1060,1067,1034,1066,1037,1047,1045,1079,1081,1048]
```

```python
path = "/Users/sicongzhao/Box/My Box Notes/Emotional-State-Capstone/data/R00/experience_sampling/Exp_Samp_individual/EmoExpSamp"
```

```python
result = pd.DataFrame()
for aid in ids:
#     print(aid)
    url = path + str(aid) + '.csv'
    exp_data = pd.read_csv(url, encoding = "ISO-8859-1")
    missing_ratio = (exp_data.isna().sum())/(exp_data.shape[0]-1)
#     print(missing_ratio)
    missing_df = missing_ratio.to_frame().T
    missing_df['id'] = aid
    result = result.append(missing_df)
del result['Unnamed: 33']
```

```python
result.to_csv('exp_missingness.csv')
```

```python
# For DND
dnd_exp_url = '/Users/sicongzhao/Box/My Box Notes/Emotional-State-Capstone/data/DND/experience_sampling/NeuromodulationDesir_DATA_2018-09-04_1916.csv'
exp_dnd= pd.read_csv(dnd_exp_url)
```

```python
exp_df = exp_dnd[['lap','hap','han','lan','la','p','n','ha','subject']]
```

```python
exp_df['subject'].unique().shape[0]
```

```python
def calc_missingness(df):
    missing_ratio = (df.isna().sum())/(df.shape[0])
#     print(missing_ratio)
    missing_df = missing_ratio.to_frame().T
    return missing_df

dnd_result = exp_df.groupby(['subject']).apply(calc_missingness)
```

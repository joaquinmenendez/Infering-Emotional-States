```python
#Append all data
for file in files[1:]:
    subj_id = file[:-4] 
    df_sleep =  pd.read_excel(path + file, sheet_name= 'Sleep')
    df_sleep['ID'] = subj_id
    df_activity =  pd.read_excel(path + file, sheet_name= 'Activities')
    df_activity['ID'] = subj_id
    #concat
    activity_dnd = pd.concat([activity_dnd, df_activity], sort=False)
    sleep_dnd = pd.concat([sleep_dnd,df_sleep], sort=False)
```

Something went wrong. It seems that the files have different structure.

Let's check this

```python
sleep_dnd.loc[sleep_dnd['Date'].isna(),:]
```

```python
prob_files = sleep_dnd.loc[sleep_dnd['Date'].isna(),:]['ID'].unique() #problematic files
prob_files
```

```python
# Subject DND007 didn't follow the 10 days in a continous way. We should drop this subject.
# The rest dont have a date column. but I can calculate this using the Start Time feature. I am making the next assumption:
# I will use the start date to set a date. If i have two start date i will add them.
for f in prob_files[1:]:
    timed = pd.to_datetime(sleep_dnd.loc[sleep_dnd['ID']== f,'Start Time'].apply(lambda x: x[:10]), format= '%Y-%m-%d')
    sleep_dnd.loc[sleep_dnd['ID']==f , 'Date'] = timed.dt.strftime('%Y-%m-%d')
```

```python
#Check if it worked
sleep_dnd.loc[sleep_dnd['Date'].isna(),:]
```

```python
#YES! let's continue
sleep_dnd.reset_index(inplace=True)
sleep_dnd.drop(columns=['index'], inplace=True)
```

```python
sleep_dnd
```

```python
def sum_dates(df_to_modif,id_subjets): 
    df = df_to_modif.copy()
    summed_df = 0
    ind = []
    for id_subj in id_subjets:
        if df['Date'].loc[df['ID'] == id_subj].shape[0] == len(df['Date'].loc[df['ID'] == id_subj].unique()):
            print('ok')
            continue
        dupli_mask = df['Date'].loc[df['ID'] == id_subj].duplicated()
        dupli = df['Date'].loc[df['ID'] == id_subj].loc[dupli_mask]
        for date in dupli:
            for i in df['Date'].loc[(df['ID'] == id_subj) & (df['Date'] == date)].index:
                ind.append(i) #filter only the repeated dates for that subject
            summed = df.loc[(df['ID'] == id_subj) & (df['Date'] == date)].sum(axis = 0)  #summing into one row
            #print(summed)
            summed[0] = date #change the date
            summed[5] = id_subj 
            summed[6:] = np.nan #remove other values
            #delete the repeated rows 
            
            if type(summed_df) != type(0) :
                summed_df = pd.concat([summed_df, pd.DataFrame(summed).T], sort= False).copy()
            else:
                summed_df = pd.DataFrame(summed).T
    return summed_df,ind
```

```python
SLEEP_DND, ind = sum_dates(sleep_dnd,prob_files[1:])
```

```python
SLEEP_DND.index = np.arange(0,SLEEP_DND.shape[0])
SLEEP_DND
```

```python
#deleting the repeated value 03-17
SLEEP_DND.drop(index=9, inplace= True)
```

```python
SLEEP_DND_Final = pd.concat([sleep_dnd,SLEEP_DND], sort = False).drop(index=ind)
```

```python
SLEEP_DND_Final.sort_values(['ID','Date'], inplace= True)
```

```python
SLEEP_DND_Final.reset_index(inplace=True, drop= True)
```

```python
SLEEP_DND_Final['Experiment'] = 'DND'
```

```python
SLEEP_DND_Final
```

# Join with DND

```python
SLEEP_R00_DND = pd.concat([r00_sleep,SLEEP_DND_Final],sort= False).reset_index(drop=True)
```

```python
with open('./All_clean_data/SLEEP_R00_DND.pickle','wb') as file:
    pickle.dump(SLEEP_R00_DND, file)
```
